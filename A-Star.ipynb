```python
class Node:
    '''
    Class node will create children for forwarded base combinations
    Helper Functions are copy, find, shuffle
    
    data: Board positions
    level: current level of node in search tree
    fval: function value response (f = g + h)
    '''
    def __init__(self, data, level, fval):
        self.data = data
        self.level = level
        self.fval = fval
        
    def copy(self,root):
        temp = []
        for i in root:
            t = []
            for j in i:
                t.append(j)
            temp.append(t)
        return temp
    
    def generateChildren(self):
        '''
        finds underscore in array
        generate possible variations
        check if variation tuples are valid 
        if valid inclulde into tree
        '''
        x,y = self.find(self.data, '_')
        shufs = [[x,y+1],[x, y-1],[x+1, y],[x-1, y]]
        children = []
        for i in shufs:
            child = self.shuffle(self.data, x, y, i[0], i[1])
            if child is not None:
                child_node = Node(child,self.level+1,0)
                children.append(child_node)
        return children
    
    def shuffle(self, puz, x1, y1, x2, y2):
        '''
        Allows valid tuples to generate nodes
        '''
        if x2 >= 0 and x2 < len(self.data) and y2 >= 0 and y2 < len(self.data):
            temp_puz = []
            temp_puz = self.copy(puz)
            temp = temp_puz[x2][y2]
            temp_puz[x2][y2] = temp_puz[x1][y1]
            temp_puz[x1][y1] = temp
            return temp_puz
        else:
            return None
        
    def find(self, puzzle, key):
        '''
        Finds _ position
        '''        
        for i in range(len(self.data)):
            for j in range(len(self.data)):
                if puzzle[i][j] == key:
                    return i,j
```

```python
class Puzzle:
    '''
    Main class 
    Accepts the arrays for board positions
    printing boards
    solver function in the puzzle
    '''
    def __init__(self, size):
        self.n = size
        self.open = []
        self.close = []

    def accept(self):
        puzzle = []
        for i in range(self.n):
            temp = input().split(" ")
            puzzle.append(temp)
        return puzzle

    def display(self, puzzle):
        print(puzzle[0])
        print(puzzle[1])
        print(puzzle[2])
    
    def f(self, start, goal):
        return self.h(start.data, goal) + start.level
    
    def h(self, start, goal):
        temp = 0
        for i in range(self.n):
            for j in range(self.n):
                if start[i][j] != goal[i][j] and start[i][j] != '_':
                    temp+=1
        return temp
    
    def solve(self):
        '''
        Solves the function based on provided start and goal postions
        Will work for 3*3, 4*4 ... board variations with solvable board poses
        '''
        print("Enter starting state:")
        start = self.accept()
        print("Enter goal state:")
        goal = self.accept()
        start = Node(start, 0, 0)
        start.fval = self.f(start, goal)
        self.open.append(start)
        while True:
            current_node = self.open[0]
            print("    ")
            self.display(current_node.data)
            if(self.h(current_node.data,goal) == 0):
                break
            for node in current_node.generateChildren():
                node.fval = self.f(node,goal)
                self.open.append(node)
            self.close.append(current_node)
            del self.open[0]
            self.open.sort(key = lambda x:x.fval,reverse=False)
```

```python
p = Puzzle(3)
```

```python
p.solve()
```

```python
p = Puzzle(4)
p.solve()
```
